## Functions

```python
# Plotting Mutual Information of 30 features against target column price on sample of 30% of data

def calculateMI(df, price_col):   
    PERCENT = 0.30
    size = df.shape[0]
    indices = np.random.choice(size, math.floor(size*PERCENT))
    x, y = df.iloc[indices,:], price_col.iloc[indices]
    y = np.array(y)
    
    scores = []
    for i, c in enumerate(x.columns):
        col = np.array(x[c]).reshape(-1,1)
        score = mutual_info_regression(col, y)
        scores.append((c, score))
        
        clear_output(wait=True)
        print(i, " | ", c, ": ", score)
    
    return scores
```
